```python
series_dictio = {}

for col in df_occ.columns:
    a_col = df_occ[col].unique()
    a_series = pd.Series(a_col)
    series_dictio[col] = a_series
    
df_occ_dos = pd.DataFrame(series_dictio)
df_occ_dos.head()
```

```python
duplicates = df_occ[df_occ.duplicated(['Occupation\ncode', 'Occupation text'])]
duplicates.head()
```

```python
dup_mask = df_occ.duplicated(['Occupation\ncode', 'Occupation text'])
```

```python
dup_mask.sum()
```

```python
to_dropit = dup_mask[dup_mask == True]
len(to_dropit)
```

```python
df_to_merge = (no_dupli.loc[no_dupli['Area level'] == 'National', 
                                    ['Occupation\ncode', 'Occupation text']]
                       .astype({'Occupation\ncode': np.int32}, copy=True))

df_to_merge.reset_index(drop=True, inplace=True)
df_to_merge.head()
```
